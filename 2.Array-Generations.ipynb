## Q.2 - Defining/Visulizing 3D-Mesh using **np.meshgrid** 
* Requirements:
  * Create a 3D-mesh as shown in the figure below
* Guidelines:
  * See the example: [https://matplotlib.org/stable/gallery/mplot3d/surface3d.html](https://matplotlib.org/stable/gallery/mplot3d/surface3d.html)

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

#YOUR CODE IS HERE
plt.figure(figsize=(100,100))
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Create the meshgrid
X = np.arange(-5, 5, 0.2)
Y = np.arange(-5, 5, 0.2)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
print(Z.max())

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis
ax.set_zlim(0, Z.max())
ax.zaxis.set_major_locator(LinearLocator(10))
ax.yaxis.set_major_locator(LinearLocator(10))
ax.xaxis.set_major_locator(LinearLocator(10))
# ax.xaxis.set_major_formatter('{x:.f}')
fig.colorbar(surf, shrink=0.5, aspect=6) # Add a color bar which maps values to colors.shrink its size to half and aspect ratio to 6 (height to width ratio)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('z = $x^2$ + $y^2$')
plt.show()

```

## Q.1 Sampling numbers from normal distribution with given **mean** and **std**
* Variables:
  * **mu**, **sigma**: are numbers defining mean and standard deviation of normal distribution

  * **N**: integer, number of samples
* Requirements:
  * Generate **N** numbers from the normal distribution 
    * NOTE: **Each time you run the code different integers are generated**
  * Show the distribution as the figure below
*Guidelines:
  * Link to wiki::normal distribution: [https://en.wikipedia.org/wiki/Normal_distribution](https://en.wikipedia.org/wiki/Normal_distribution)
  * Link to numpy API: [https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.htmlt](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
# Declare variables
mu, sigma = 5, 3
N = [100, 1000, 10000, 100000, 1000000]

#YOUR CODE IS HERE\
fig, ax = plt.subplots(3,2, figsize=(10, 30))
for i, n in enumerate(N):
    x = np.random.normal(mu, sigma, n)
    x_norm = np.linspace(mu - 3*sigma, mu + 3*sigma, n)
    pdf_vals = norm.pdf(x_norm, mu, sigma)
    ax[i//2, i%2].hist(x, bins=100, density=True )
    ax[i//2, i%2].plot(x_norm, pdf_vals, c='r', linewidth=2)
    ax[i//2, i%2].set_title(f"Histogram for {n} points")
```
