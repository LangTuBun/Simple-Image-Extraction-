# 1- Slicing - (I)

## 1.1- Examples
* Syntax:
  * use ```':'``` to select all indices at one dimension
  * use ```'...'``` to select all the indices in all the remaining dimensions

````python
# declare variables
N, M = 6, 4

# generate N integers
np.random.seed(0)
X = np.random.randint(0, 10, (N,M))

# select items using their indices
np.set_printoptions(formatter={'float': '{:<0.3f}'.format, 
                                'int': '{:<3d}'.format,
                                'bool': '{:<3}'.format })
print("X: \t"); print(X)
print('-'*150)

print('To select colum 0, X[:, 0]: ', X[:, 0])
print('To select colum 0, X[..., 0]: ', X[..., 0])
print('To select colum 2, X[:, 2]: ', X[:, 2])
print('To select colum 2, X[..., 2]: ', X[..., 2])
print('-'*50)
print('To select colum 0 and 2, X[:, [0, 2]]: '); print(X[:, [0,2]])
print('To select colum 0 and 2, X[:, (0, 2)]: '); print(X[:, (0,2)])
print('To select colum 0 and 2, X[..., [0, 2]]:'); print(X[..., [0,2]])
print('To select colum 0 and 2, X[..., (0, 2)]:'); print(X[..., (0,2)])
````

* Information:

    * size of axis 0 (or -3): 6

* Requirements:
  1. Select items satisfy: 
    * axis 0: select indices [1,3,0]

````python
# declare variables
N, M, K = 6, 5, 3

# generate N integers
np.random.seed(0)
X = np.random.randint(0, 10, (N,M,K))

# select items using their indices
np.set_printoptions(formatter={'float': '{:<0.3f}'.format, 
                                'int': '{:<3d}'.format,
                                'bool': '{:<3}'.format })
print("X: \t"); print(X)
print('-'*150)

#YOUR CODE IS HERE
print('axis 0: select indices [1,3,0]\n', X[[1,3,0], :,:])
print('axis 0 select indices [1,3,0]\naxis 2: select indices [2,1]\n', X[[1,3,0],...][:,:, [2,1]])
print('axis 1: select indices [2,1,4]\n', X[:, [2,1,4]])

````
